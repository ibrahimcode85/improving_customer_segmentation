## MAIN BUSINESS QUESTIONS:
- # What features is a good predictor to determine if a doctor will proceed with purchase or not (binary) or the number of purchase (regression).
- # If doctors can be segmentized based on this predictors.

```python
# import required libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# load dataset
master_df = pd.read_csv(r".\data\master_df_pprocess1.csv")
```

```python
numeric_summary = master_df.describe()
display(numeric_summary)
```

```python
# Check count of missing values for each column
missing_values_count = master_df.isna().sum()

# Display the count of missing values
print(missing_values_count)
```

```python
non_numeric_summary = {col: master_df[col].unique() for col in master_df.columns if master_df[col].dtype == 'object'}
for col, unique_values in non_numeric_summary.items():
    print(f"Column: {col}")
    print(f"Unique Values: {unique_values}\n")
```

### Judgement Made!
For `Satisfaction` column - the score contains value '--' which we suspect the doctor is not participating in giving a satisfaction rating. For practicality - we will give neutral score of 0.

For `Rank` column - there are two missing value. Replace that with the most occurence value of that column.

```python
# Replace '--' with 0 in the 'Satisfaction' column and convert to numeric
master_df['Satisfaction'] = master_df['Satisfaction'].replace('--', 0).astype(float)

# Get summary statistics for the 'Satisfaction' column
satisfaction_summary = master_df['Satisfaction'].describe()
print(satisfaction_summary)
```

```python
# Find the most frequent value in the 'Rank' column
most_frequent_rank = master_df['Rank'].mode()[0]

# Replace missing values in the 'Rank' column with the most frequent value
master_df['Rank'].fillna(most_frequent_rank, inplace=True)

# Verify that there are no more missing values in the 'Rank' column
missing_values_count = master_df.isna().sum()
print(missing_values_count)
```

```python
# Count the number of unique region codes
unique_regions = master_df['Region'].nunique()

print(f"Number of unique region codes: {unique_regions}")


# Label encode the 'Region' column
label_encoder = LabelEncoder()
master_df['Region'] = label_encoder.fit_transform(master_df['Region'])
```

```python
# Convert 'Yes' to 1 and 'No' to 0 in the 'Instructions' column
master_df['Instructions'] = master_df['Instructions'].map({'Yes': 1, 'No': 0})
```

```python
master_df.to_csv(r".\data\master_df_final.csv", index=False)
```
